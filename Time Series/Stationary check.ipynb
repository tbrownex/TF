#### Use the Dickey Fuller test for stationarity

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
def checkADF(data):
    return adfuller(data)
```

##### Generate a time series with some Trend

```python
size=300
noise = np.random.normal(size=size)
init = 1.0
Beta = 0.98      # Smaller Beta means more stationary

series = [init]
for x in range(1,size):
    tmp = series[x-1]*Beta + noise[x]
    series.append(tmp)

plt.plot(series)
plt.show()
```

```python
result = checkADF(series)
ADFstat = result[0]
Pvalue  = result[1]
print('ADF Statistic: {:.3f}'.format(ADFstat))
print('P-value: {:>15.6f}'.format(Pvalue))

print("\n", "Confidence Intervals")
for key, value in result[4].items():
    if ADFstat < value:
        txt = "Stationary"
    else:
        txt = "Non-stationary"
    print("{:<5}{:<10.3f}{}".format(key, value, txt))
```
