```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
weeks= 230
base = 10000

original = []
for x in range(weeks):
    tmp = base + 1000 * np.random.normal()
    if x % 12 == 0:
        tmp += 4000
    original.append(tmp)

original = np.array(original)
plt.plot(original)
plt.ylim(0)
plt.show()
```
